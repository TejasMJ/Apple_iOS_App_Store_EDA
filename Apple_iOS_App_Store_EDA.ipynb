#### Based on the publicly available iOS App Store dataset, the meaning of each column is as follows:

- id: Unique identifier of the app in the App Store.
- track_name: Name of the application as listed on the App Store.
- size_bytes: Size of the app in bytes.
- currency: Currency type used for the app price (e.g., USD).
- price: Price of the app (0 indicates a free app).
- rating_count_tot: Total number of user ratings across all versions of the app.
- rating_count_ver: Number of user ratings for the current version of the app.
- user_rating: Average user rating for all versions of the app.
- user_rating_ver: Average user rating for the current version of the app.
- ver: Latest version code of the app.
- cont_rating: Content rating indicating target age group suitability (e.g., 4+, 12+, 17+).
- prime_genre: Primary genre/category of the app.
- sup_devices.num: Number of devices supported by the app.
- ipadSc_urls.num: Number of screenshots available for display on iPad.
- lang.num: Number of supported languages in the app.
- vpp_lic: Indicates if VPP (Volume Purchase Program) device-based licensing is enabled.

#### Checking all unique values of each column

```python
# Count of NaN / None values per column
nan_count = df.isnull().sum()

# Count of empty strings per column
empty_count = (df == '').sum()

# Total missing values (NaN + empty strings)
total_missing = nan_count + empty_count

# Percentage of missing values
missing_percent = (total_missing / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Values': total_missing,
    'Percentage (%)': missing_percent
}).sort_values(by='Percentage (%)', ascending=False)

print(missing_df)
```

## Data Cleaning

#### Removing the Duplicate records in track_name                     

```python
total_ids = df.shape[0]
unique_names = df['track_name'].nunique()
duplicate_names = df[df.duplicated(subset=['track_name'], keep=False)]['track_name'].unique()

print("=== Duplicate Check ===")
print(f"Total app IDs: {total_ids}")
print(f"Unique app names: {unique_names}")
print(f"Number of duplicate app names: {len(duplicate_names)}")
print(f"Duplicate app names: {duplicate_names}")

# ------------------------------
# Step 2: Remove duplicates
# ------------------------------
df_clean = df.drop_duplicates(subset=['track_name'], keep='first')
df_clean.reset_index(drop=True, inplace=True)

print("\n=== After Removing Duplicates ===")
print(f"Original dataset shape: {df.shape}")
print(f"Dataset shape after removing duplicates: {df_clean.shape}")
```

#### Converting byte size to mega bytes (MB)

```python
# Convert size_bytes to MB
df['size_mb'] = (df['size_bytes'] / (1024 * 1024)).round(2)

# Check the conversion
print(df[['track_name', 'size_bytes', 'size_mb']].head())
```

#### Removing Unwanted Columns

```python
# Columns you want to remove
columns_to_remove = ['Unnamed: 0', 'id', 'size_bytes', 'vpp_lic']  # example columns

# Remove columns
df = df.drop(columns=columns_to_remove)
```

#### Univaraite Analysis

```python
# =====================================================
# UNIVARIATE ANALYSIS
# =====================================================

# -------------------------
# CATEGORICAL VARIABLES
# -------------------------

# 1 Number of Apps per Genre (All Genres, Sorted, Horizontal)

genre_counts = (
    df['prime_genre']
    .value_counts()
    .rename_axis('Genre')
    .reset_index(name='Count')
)

fig = px.bar(
    genre_counts,
    x='Count',
    y='Genre',
    orientation='h',
    title='Number of Apps per Genre',
    text='Count'
)

fig.update_traces(textposition='outside')
fig.update_layout(
    yaxis=dict(categoryorder='total ascending', automargin=True),
    height=900
)

fig.show()


# 2 Free vs Paid Apps Distribution

# Create Free vs Paid column
df['Type'] = df['price'].apply(lambda x: 'Free' if x == 0 else 'Paid')

type_counts = (
    df['Type']
    .value_counts()
    .rename_axis('Type')
    .reset_index(name='Count')
)

fig = px.bar(
    type_counts,
    x='Type',
    y='Count',
    title='Free vs Paid Apps',
    text='Count',
    color='Type'
)

fig.update_traces(textposition='outside')
fig.show()


# 3 Content Rating Distribution

content_counts = (
    df['cont_rating']
    .value_counts()
    .rename_axis('Content Rating')
    .reset_index(name='Count')
)

fig = px.bar(
    content_counts,
    x='Content Rating',
    y='Count',
    title='Apps by Content Rating',
    text='Count'
)

fig.update_traces(textposition='outside')
fig.show()
```

#### Multivariate Analysis

```python
# -------------------------
# Function to style box plots
# -------------------------
def style_box(fig, xaxis_title, yaxis_title, title_text):
    fig.update_traces(
        boxmean='sd'  # shows mean and standard deviation
    )
    
    fig.update_layout(
        width=1500,
        height=800,
        xaxis_title=dict(text=xaxis_title, font=dict(size=25, color='black')),
        yaxis_title=dict(text=yaxis_title, font=dict(size=25, color='black')),
        xaxis=dict(
            showgrid=False,
            tickfont=dict(size=18, color='black'),
            tickangle=45
        ),
        yaxis=dict(showgrid=False, tickfont=dict(size=20, color='black')),
        title=dict(
            text=title_text,
            font=dict(size=25, color='black'),
            x=0.5,
            xanchor='center'
        ),
        plot_bgcolor='rgba(0,0,0,0)',
        paper_bgcolor='rgba(0,0,0,0)',
        legend=dict(
            title='Type',
            font=dict(size=18, color='black'),
            title_font=dict(size=20, color='black')
        )
    )
    
    return fig

# -------------------------
# Filter top 15 genres
# -------------------------
top15_genres = df['prime_genre'].value_counts().nlargest(15).index
df_top15 = df[df['prime_genre'].isin(top15_genres)]

# -------------------------
# Box plot: User Rating by Genre and Type
# -------------------------
fig = px.box(
    df_top15,
    x='prime_genre',
    y='user_rating',
    color='Type',
    color_discrete_sequence=['#43aa8b', '#f94144'],
    title='User Rating by Genre and Type (Top 15 Genres)'
)

# Apply styling
fig = style_box(
    fig,
    'Genre',
    'User Rating',
    'User Rating by Genre and Type (Top 15 Genres)'
)

fig.show()
```

```python
import plotly.express as px

# -------------------------
# Function to style histograms
# -------------------------
def style_histogram(fig, xaxis_title, yaxis_title, title_text):
    # Apply text styling only to histogram traces
    for trace in fig.data:
        if trace.type == 'histogram':
            trace.update(
                textposition='outside',
                textfont=dict(color='black', size=25)
            )
    
    # Layout styling
    fig.update_layout(
        width=1500,
        height=800,
        xaxis_title=dict(text=xaxis_title, font=dict(size=25, color='black')),
        yaxis_title=dict(text=yaxis_title, font=dict(size=25, color='black')),
        xaxis=dict(showgrid=False, tickfont=dict(size=20, color='black')),
        yaxis=dict(showgrid=False, tickfont=dict(size=20, color='black')),
        title=dict(
            text=title_text,
            font=dict(size=25, color='black'),
            x=0.5,
            xanchor='center'
        ),
        plot_bgcolor='rgba(0,0,0,0)',
        paper_bgcolor='rgba(0,0,0,0)',
        uniformtext_minsize=8,
        uniformtext_mode='hide',
        showlegend=False
    )
    
    return fig

# -------------------------
# 1. Total Rating Count Distribution
# -------------------------
fig = px.histogram(
    df,
    x='rating_count_tot',
    nbins=50,
    text_auto=True,
    color_discrete_sequence=['#aa2927'],
    marginal='box'
)
fig = style_histogram(fig, 'Reviews', 'Number of Apps', 'Distribution of Reviews')
fig.show()

# -------------------------
# 2. App Size Distribution (MB)
# -------------------------
fig = px.histogram(
    df,
    x='size_mb',
    nbins=50,
    text_auto=True,
    color_discrete_sequence=['#fdbf4a'],
    marginal='box'
)
fig = style_histogram(fig, 'Size (MB)', 'Number of Apps', 'Distribution of App Size')
fig.show()

# -------------------------
# 3. User Rating vs Total Rating Count (Scatter)
# -------------------------
fig = px.scatter(
    df,
    x='rating_count_tot',
    y='user_rating',
    opacity=0.3,
    color_discrete_sequence=['#548d8a'],
    title='User Rating vs Total Rating Count'
)

# Layout styling for scatter
fig.update_layout(
    width=1500,
    height=800,
    xaxis_title='Total Rating Count',
    yaxis_title='User Rating',
    xaxis=dict(showgrid=False, tickfont=dict(size=20, color='black')),
    yaxis=dict(showgrid=False, tickfont=dict(size=20, color='black')),
    plot_bgcolor='rgba(0,0,0,0)',
    paper_bgcolor='rgba(0,0,0,0)',
    title=dict(font=dict(size=25, color='black'), x=0.5, xanchor='center')
)
fig = style_histogram(fig, 'Reviews', 'Ratings', 'Rating vs Reviews (Normal Scale)')
fig.show()

# Filter paid apps
paid_apps = df[df['price'] > 0]

# Create histogram with styling similar to your example
fig = px.histogram(
    paid_apps,
    x='price',
    nbins=30,
    text_auto=True,
    color_discrete_sequence=['#548d8a'],
    marginal='box',
    title='Price Distribution (Paid Apps)'
)

# Apply your styling function
fig = style_histogram(fig, 'Price', 'Number of Apps', 'Price Distribution (Paid Apps)')

fig.show()
```
